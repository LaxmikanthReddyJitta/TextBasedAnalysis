```python
# Printing the value of the string variable 'name'
name="kanth"
print(name)
```

```python
# Using slice( method) to extract a substring from the string 'name' (characters at index 1 to 2)
print(name[1:3])
```

```python
print(name.upper())  #using upper() method to convert string to upper cases
print(name.lower())#using lower() method to convert string to lower cases
print(name.replace("t","T"))#using replace() method to replace one character with other in a string
```

```python
# Concatenating the strings 'frst' and 'name' to create a new string 'fullName'
frst='laxmi'
fullName=frst+name
print(fullName)
```

```python
#using for loop to print each element in the string
for i in fullName:
  print(i)
```

```python
# Iterating through each index in the range from 0 to the length of the string 'fullName'
for i in range (0,len(fullName)):
  # Printing the character at the current index in 'fullName'
  print(fullName[i])

```

```python
# Initializing a string variable with space separated values
myString="1 2 3 4 5"
# Using the split() method to separate the string into a list of substrings according to whitespace
print(myString.split())
```

```python
# Initializing a string variable with values separated by forward slash
myString="1/2/3/4/5"
# Using the split("/") method to separate the string into a list of substrings based on the forward slash delimiter
print(myString.split("/"))
```

```python
# Initializing a string variable with values separated by the "@" symbol
myString="1@2@3@4@5"
# Using the split("@") method to separate the string into a list of substrings based on the "@" delimiter
print(myString.split("@"))
```

```python
import re #importing regular expression module
# Initializing a string variable with mixed values separated by "/" and "@"
myString="1@2@3/4/5"
# Using re.split() with a regular expression to split the string into a list of substrings based on "/" or "@"
print(re.split("@|/" , myString))
```

here the special characters are called **DELIMETERS**


```python
# Using the find() method to locate the index of the substring "kanth" in the string 'fullName'
print(fullName.find("kanth"))
```

```python
# Using the startswith() method to check if the string 'fullName' starts with the substring "kanth"
print(fullName.startswith("kanth"))
```

```python
# Using the lstrip() method to remove leading whitespaces from the string 'language'
language="   python"
print(language.lstrip())
```

```python
 #Using the lstrip() method with the argument "p" to remove leading occurrences of "p" from the string 'language'
print(language.lstrip("p"))
```

```python
# Using the strip() method to remove leading and trailing whitespaces from the string 'language'
print(language.strip())
```
