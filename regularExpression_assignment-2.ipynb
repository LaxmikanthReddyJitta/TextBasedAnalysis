**REGULAR EXPRESSIONS**

REGEX :Regular expression (regex or regexp) is a sequence of characters that forms a search pattern. It is a powerful tool for string manipulation and searching. The re module in Python provides support for regular expressions.

```python
# importing regex package
import regex as re

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

**TOKENIZATION**

```python
#generating tokens
testtext = "This data set is about a article of newyork times"
testtokens = tokenize(testtext)

print(testtokens)
```


**STOPWORDS**

In natural language processing, stopwords are commonly used words (such as "the," "is," "and," etc.) that are often excluded from text processing because they do not carry significant meaning and are used for grammatical structure. In Python, the nltk (Natural Language Toolkit) library is commonly used for working with text data, including handling stopwords.

```python
#importing nltk package
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

**TOKENIZATION**

```python
# Removing stopwords from given string
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```


**TEXT PROCESSING PIPELINE**

```python
#Lowercasing, Tokenization, and Stopword Removal
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

**TEXT PROCESSING PIPELINE**


```python
#Applying a custom text processing pipeline to the 'text' column in the DataFrame
df['tokens'] = df['headline'].apply(prepare, pipeline=pipeline)
```

**COUNTING WORDS**

A Counter is a data type in the collections module. It is a dict subclass that stores counts for hashable objects.

```python
from collections import Counter

tokens = tokenize("doing regular exercises and eating good food keep us healthy every time")

testcount= Counter(tokens)   #counts each word and its occurence
print(testcount)
```

**DISPLAYING KEY VALUES**

```python

#iterates over each entry,updating the counter with the token counts
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

**CREATING DATAFRAME AND FILTERING ENTRIES**

```python
#include only rows where the 'freq' column is greater than or equal to 1.
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

**SORTING**

```python
#Sorts and displays
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

**WORDCLOUD**

Word cloud is one of the most powerful and straightforward visualization methods when it comes to text data. The size of words are dependent on the occurrence frequency

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load mask image
mask = np.array(Image.open("images.png"))

# Convert mask image to RGB if it's in grayscale
if len(mask.shape) == 2:
    mask = np.stack((mask,) * 3, axis=-1)

# Create WordCloud object with mask
wc = WordCloud(background_color='white', mask=mask, contour_color='steelblue', contour_width=1, max_words=100)

# Assuming freq_df['freq'] contains frequency of words
freq_dict = freq_df['freq'].to_dict()

# Generate word cloud
wordcloud = wc.generate_from_frequencies(freq_dict)

# Create coloring from image
image_colors = ImageColorGenerator(mask)

# Display the generated image
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()



```
