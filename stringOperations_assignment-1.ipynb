**capitalize():**

The capitalize() method returns a string where the first character is upper case, and the rest is lower case.

```python
# Define a variable 'course' and assign it the value "text based analysis"
course = "text based analysis"

# Call the 'capitalize()' method on the 'course' string, which capitalizes the first letter of the string
x = course.capitalize()

# Print the capitalized string
print(x)
```

**casefold():**

The casefold() method returns a string where all the characters are lower case, This method is similar to the lower() method, but the casefold() will convert more characters into lower case

```python
# Define a variable 'course' and assign it the value "Text Based Analysis"
course = "Text Based Analysis"

# Call the 'casefold()' method on the 'course' string, which converts all characters in the string to lowercase
x = course.casefold()

# Print the lowercase version of the string
print(x)
```

**find():**

The find() method finds the first occurrence of the specified value.
The find() method returns -1 if the value is not found.

```python
# Define a variable 'course' and assign it the value "Text Based Analysis"
course = "Text Based Analysis"

# Call the 'find()' method on the 'course' string, which searches for the specified substring ("Analysis") within the string
# It returns the lowest index in the string where the substring is found, or -1 if the substring is not found
x = course.find("Analysis")

# Print the index where the substring "Analysis" starts within the string
print(x)
```

**format():**

The format() method formats the specified value(s) and insert them inside the string's placeholder.

```python
# Define a variable 'course' and assign it the value "the class is {clas:}"
course = "the class is {clas:}"

# Use the 'format()' method on the 'course' string to replace the placeholder '{clas:}' with the provided value
formatted_course = course.format(clas="text based analysis")

# Print the formatted string
print(formatted_course)
```

**isdecimal()**

The isdecimal() method returns True if all the characters are decimals (0-9).

```python
# Define a variable 'num' and assign it the value "1234"
num = "1234"

# Call the 'isdecimal()' method on the 'num' string, which checks if all characters in the string are decimal digits (0-9)
x = num.isdecimal()

# Print the result, which indicates whether all characters in the string are decimal digits
print(x)
```

**join()**

The join() method takes all items in an iterable and joins them into one string.

```python
# Define a tuple 'names' containing three strings: "laxmi", "kanth", "reddy"
names = ("laxmi", "kanth", "reddy")

# Use the 'join()' method to concatenate all elements of the 'names' tuple into a single string
# The "#" character is used as a separator between the elements
x = "#".join(names)

# Print the resulting string
print(x)
```

**replace()**

The replace() method replaces a specified phrase with another specified phrase

```python
# Define a string variable 'name' containing the value "laxmi kanth jitta"
name = "laxmi kanth jitta"

# Use the 'replace()' method to replace all occurrences of the substring "jitta" with the substring "reddy"
x = name.replace("jitta", "reddy")

# Print the resulting string after replacement
print(x)
```

**partition()**

The partition() method searches for a specified string, and splits the string into a tuple containing three elements.



```python
# Define a string variable 'name' containing the value "laxmianth reddy jitta"
name = "laxmianth reddy jitta"

# Use the 'partition()' method to partition the string at the first occurrence of the specified separator "jitta"
x = name.partition("jitta")

# Print the resulting tuple
print(x)
```

**swapcase()**

The swapcase() method returns a string where all the upper case letters are lower case and vice versa

```python
# Define a string variable 'name' containing the value "laxmianth reddy jitta"
name = "laxmianth reddy jitta"

# Use the 'partition()' method to partition the string at the first occurrence of the specified separator "jitta"
x = name.partition("jitta")

# Print the resulting tuple
print(x)
```
