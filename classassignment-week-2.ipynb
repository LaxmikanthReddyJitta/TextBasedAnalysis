REGULAR EXPRESSIONS

```python
import regex as re #importing regex module.
def tokenize(text):
  return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
#storing the string in testtext variable
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

STOPWORDS

```python
#importing the nltk module
import nltk
```

```python
#printing the stopwords using nltk module
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
#creating the function for removing stop words
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
```

```python
#calling method by giving tetstokens as input
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

```python
#creating the pipeline
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
#running the pipeline
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

```python
#counting the tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

```python
#importing the counter package from collections package
from collections import Counter
tokens = tokenize("She likes my cats and my cats like my sofa.")
testcount= Counter(tokens)
print(testcount)
```

```python
#adding new data to current count
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

```python
#counting the words
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
#put the word frequency in  owndataframe for analysis
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
#sorting and displaying words and its frequencies
freq_df = freq_df.sort_values('freq', ascending=False) # sorting according to frequency
freq_df.head(10)
```

```python
#statistical analysis of frq_df dataframe
freq_df.describe()
```
